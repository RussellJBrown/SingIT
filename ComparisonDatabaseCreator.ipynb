```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.signal as signal
import librosa
import librosa.display
import librosa.feature
import IPython.display as ipd
import os
from timeit import default_timer as timer
```

```python
def to_str(var):
    return str(list(np.reshape(np.asarray(var), (1, np.size(var)))[0]))[1:-1]
```

```python
def getListOfFiles(dirName):
    # create a list of file and sub directories
    # names in the given directory
    listOfFile = os.listdir(dirName)
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)

    return allFiles
    
```

```python
#This process will go through
#Train All Songs
start = timer()
n = 10
filtertrain = [[0.0]]*13
filtertrain[0] = signal.butter(n, [3000,4000], 'bandpass', fs=44100, output='sos')
filtertrain[1] = signal.butter(n, [4000,5000], 'bandpass', fs=44100, output='sos')
filtertrain[2] = signal.butter(n, [5000,6000], 'bandpass', fs=44100, output='sos')
filtertrain[3] = signal.butter(n, [6000,7000], 'bandpass', fs=44100, output='sos')
filtertrain[4] = signal.butter(n, [7000,8000], 'bandpass', fs=44100, output='sos')
filtertrain[5] = signal.butter(n, [8000,9000], 'bandpass', fs=44100, output='sos')
filtertrain[6] = signal.butter(n, [9000,10000], 'bandpass', fs=44100, output='sos')
filtertrain[7] = signal.butter(n, [10000,11000], 'bandpass', fs=44100, output='sos')
filtertrain[8] = signal.butter(n, [11000,12000], 'bandpass', fs=44100, output='sos')
filtertrain[9] = signal.butter(n, [12000,13000], 'bandpass', fs=44100, output='sos')
filtertrain[10] = signal.butter(n, [13000,14000], 'bandpass', fs=44100, output='sos')
filtertrain[11] = signal.butter(n, [14000,15000], 'bandpass', fs=44100, output='sos')
filtertrain[12] = signal.butter(n, 15000, 'highpass', fs=44100, output='sos')
heightVar = 0.375
distanceVar = 4000

songArrayE = getListOfFiles("/media/russell/775C-44EC/outputEnglish30")
songArrayF = getListOfFiles("/media/russell/775C-44EC/outputFrench30") 
songArrayG = getListOfFiles("/media/russell/775C-44EC/outputGerman30")
songArrayJ = getListOfFiles("/media/russell/775C-44EC/outputJapanese30")

englishPeaks = "EnglishPeaks30.txt"
frenchPeaks = "FrenchPeaks30.txt"
germanPeaks = "GermanPeaks30.txt"
japanesePeaks = "JapanesePeaks30.txt"

occurenceMatrix = [[0]*len(filtertrain)]*90
count = 0
for i in range(0,180):
    if "vocals.wav" in songArrayE[i]:
        y, sr = librosa.load(songArrayE[i])
        f = open(englishPeaks,'a')
        for j in np.arange(len(filtertrain)):
            filtered = signal.sosfilt(filtertrain[j], y)
            ypeaks = signal.find_peaks(filtered, height=np.max(filtered)*heightVar, distance=distanceVar)
            occurenceMatrix[count][j] = len(ypeaks[0])                
            f.write(to_str(occurenceMatrix[count][j]))
            f.write(" ")
        count+=1
        f.write("\n")
        f.close()    
                
occurenceMatrix = [[0]*len(filtertrain)]*90        
count=0
for i in range(0,180):
    if "vocals.wav" in songArrayF[i]:
        y, sr = librosa.load(songArrayF[i])
        f = open(frenchPeaks,'a')
        for j in np.arange(len(filtertrain)):
            filtered = signal.sosfilt(filtertrain[j], y)
            ypeaks = signal.find_peaks(filtered, height=np.max(filtered)*heightVar, distance=distanceVar)
            occurenceMatrix[count][j] = len(ypeaks[0])       
            f.write(to_str(occurenceMatrix[count][j]))
            f.write(" ")
        count+=1
        f.write("\n")
        f.close()
        
                
occurenceMatrix = [[0]*len(filtertrain)]*90  
count=0
for i in range(0,180):
    if "vocals.wav" in songArrayG[i]:
        y, sr = librosa.load(songArrayG[i])
        f = open(germanPeaks,'a')
        for j in np.arange(len(filtertrain)):
            filtered = signal.sosfilt(filtertrain[j], y)
            ypeaks = signal.find_peaks(filtered, height=np.max(filtered)*heightVar, distance=distanceVar)
            occurenceMatrix[count][j] = len(ypeaks[0])        
            f.write(to_str(occurenceMatrix[count][j]))
            f.write(" ")
        count+=1
        f.write("\n")
        f.close()
        
                
occurenceMatrix = [[0]*len(filtertrain)]*90  
count=0
for i in range(0,180):
     if "vocals.wav" in songArrayJ[i]:
        y, sr = librosa.load(songArrayJ[i])
        f = open(japanesePeaks,'a')
        for j in np.arange(len(filtertrain)):
            filtered = signal.sosfilt(filtertrain[j], y)
            ypeaks = signal.find_peaks(filtered, height=np.max(filtered)*heightVar, distance=distanceVar)
            occurenceMatrix[count][j] = len(ypeaks[0])
            f.write(to_str(occurenceMatrix[count][j]))
            f.write(" ")
        count+=1
        f.write("\n")
        f.close()
        
end = timer()        
print(end-start)
```
